```python
import pandas as pd
import numpy as np
df=pd.read_csv('college.csv')
df
```

```python
df.info()
```

```python
x=df.iloc[:,1:18].values
x
```

```python
#2.Use LabelEncoder to encode the target variable:
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.iloc[:,0]=le.fit_transform(df.iloc[:,0])
df
```

```python
y=df.iloc[:,0].values
y
```

```python
#Split the data set into training and testing data:
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
len(x_test)
```

```python
#3.Fit a linear svm from scikit learn and observe the accuracy:
#SVM:
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
obj=SVC(kernel='linear')
linear_pred=obj.fit(x_train,y_train).predict(x_test)
Accuracy_svm=accuracy_score(y_test,linear_pred)
Accuracy_svm
```

```python
#4.Preprocess the data using StandardScalar:
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X=ss.fit_transform(x)
X
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
len(y_test)
```

```python
#SVM:
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
ss=SVC(kernel='linear')
linear_pred=ss.fit(x_train,y_train).predict(x_test)
Accuracy_svm=accuracy_score(y_test,linear_pred)
Accuracy_svm
```

```python
#5. Use scikit learn’s gridsearch to select the best hyperparameter for a non-linear SVM,
#identify the model with best score and its parameters:
from sklearn.model_selection import GridSearchCV
para=[{'kernel':['rbf','sigmoid']}]
GS=GridSearchCV(estimator=ss,param_grid=para,scoring='accuracy')
GS.fit(x_train,y_train)
print(GS.best_score_,GS.best_params_)
```
